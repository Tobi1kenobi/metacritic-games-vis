```python
##Load data about games
meta_games = pd.read_csv('../Data/extra_details_complete.csv')

# make new year column and replace tbd values with np.nans
meta_games['year'] = np.nan
meta_games.replace('tbd', np.nan)

# enter year value from date column
for i, row in meta_games.iterrows():
    year_value = meta_games['date'][i].split(',')[1].strip()
    meta_games.at[i,'year'] = int(year_value)

# replace tbd values with nan    
meta_games = meta_games.replace('tbd', np.nan)
meta_games_years = {}

# make a dict of meta_game dataframes for each year from 2000 to 2019
for i in range(2000,2019):
    meta_games_years[i] = meta_games[(meta_games['year'] == i)]

# remove tbd and replace with nan
meta_games = meta_games.replace('tbd', np.nan)

# make dict for consoles
console_dict = {}
for console in np.unique(meta_games['console']):
    console_dict[console] = meta_games[(meta_games['console'] == console)] 
    
# get metascore for pc
# consoles_dict['PC']['metascore']
#len(consoles_dict['PC']['userscore'].astype(float)*10 ) > 20

# make dict for publishers
pub_dict = {}
for pub in np.unique(meta_games['publisher']):
    all_by_pub = meta_games[(meta_games['publisher'] == pub)]
    if len(all_by_pub) > 10:
        pub_dict[pub] = meta_games[(meta_games['publisher'] == pub)]
    
#console_dict['PC']['metascore']
#pub_dict['Electronic Arts']['metascore']

#remove wrong pc and vita console
del console_dict[' PC']
del console_dict[' VITA']
#console_dict.keys()
#print userscores for pc
#console_dict['PC']['userscore'].astype(float)*10
```

```python
import plotly.plotly as py
import plotly.graph_objs as go 

from datetime import datetime
import pandas as pd

import plotly
plotly.tools.set_credentials_file(username='thomasnicolet', api_key='ln0vpFkn0fVIxOkBwpnL')



df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv')
df.columns = [col.replace('AAPL.', '') for col in df.columns]

#proper histogram with dropdown menu


#make data traces of metascore, one for each console, make data a list of these hist traces.
data_console=[]
for console in console_dict:
    console_metatrace = go.Histogram(
                x=console_dict[console]['metascore'],
            ##ADD/REMOVE HISTNORM FOR NORMALISING
            histnorm='percent',
            name='metascore for '+str(console),
            xbins=dict(
            start=0,
            end=100,
            size=5
            ),
            marker=dict(
            color='#8181F7',
            ),
            opacity=0.75
    )
    console_usertrace = go.Histogram(
                x=console_dict[console]['userscore'].astype(float)*10,
            ##ADD/REMOVE HISTNORM FOR NORMALISING
            histnorm='percent',
            name='userscore for '+str(console),
            xbins=dict(
            start=0,
            end=100,
            size=5
            ),
            marker=dict(
            color='#FFD7E9',
            ),
            opacity=0.75
    )
    data_console.append(console_metatrace)
    data_console.append(console_usertrace)


#make list of lists of bools, e.g. [True, False, False], [False, True, False], [False, False, True]
#used in args within buttons
console_bool_list1 = []
for i in range(0, len(console_dict)*2, 2):
    temp = [False]*(len(console_dict)*2)
    temp[i] = True
    temp[i+1] = True
    console_bool_list1.append(temp)

#making list of dictionaries to be used as argument for buttons in updatemenus
buttons_list=[]
for i in list(range(0, len(console_dict))):
    temp_list = dict(label = list(console_dict)[i],
                    method = 'update',
                    args = [{'visible': console_bool_list1[i]},
                {'title': list(console_dict)[i]}])
    buttons_list.append(temp_list)

#make update menus. buttons is made up of temp_list, i.e. list of dictionaries containing label, method, args and title. 
#args variable determines when what is being shown, e.g. [True, False False... False] means first is being shown only
updatemenus = list([
    dict(active=1,
         buttons=buttons_list
    )
])

layout = dict(title='Metascores for platforms etc.', 
              showlegend=False, 
              yaxis = dict(
                  range = [0, 30]
              ),
              xaxis = dict(
                  range = [0, 100]
              ),
              updatemenus= updatemenus)

fig = dict(data=data_console, layout=layout)
py.iplot(fig, filename='update_dropdown')
```
